```python
from bs4 import BeautifulSoup as bs
import requests

url = 'https://buildinfos.debian.net/ftp-master.debian.org/buildinfo/2021/01/18/'

page = requests.get(url)

soup = bs(page.content,"html.parser")      # beautiful soup for reading the webpage



anchors = soup.find_all('a')


all_links = []
for link in anchors:
    linkText = link.get('href') #find the html reference text mentioned between the anchor tags
    all_links.append(linkText)
```

```python
all_links = all_links[5:]

for i in all_links:
    url = 'https://buildinfos.debian.net/ftp-master.debian.org/buildinfo/2021/01/18/' + i
    page = requests.get(url)

    soup = bs(page.content,"html.parser")

    with open(i,'w') as f:   # save the text file from the website
        f.write(str(soup))
```

```python
A = {}
for i in all_links:
    D = []
    with open(i) as f:
        for line in f:
            if ('(' in line):
                temp = line.strip()
                D.append(temp)
        node = i.replace('.buildinfo','')
        A[node] = D
```
